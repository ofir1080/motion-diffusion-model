```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Example sequence of joints
joints = np.load('test.npy', allow_pickle=True)
x = [joint[:, 0] for joint in joints[99:100]]
y = [joint[:, 1] for joint in joints[99:100]]
z = [joint[:, 2] for joint in joints[99:100]]

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x, y, marker='o')  # 'marker' adds a point at each joint
ax.view_init(elev=90, azim=-90)  
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Sequence of Joints in 3D')

plt.show()
```
